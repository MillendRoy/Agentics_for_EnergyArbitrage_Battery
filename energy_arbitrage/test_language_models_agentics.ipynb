```python
import numpy as np
import matplotlib.pyplot as plt

from agentic_energy.language_models.basic_llm_transduction import client_calling

result,day,battery_details = await client_calling()
```

```python
output =result.states[0]
output
```

```python
day
```

```python
import matplotlib.pyplot as plt
hours = np.arange(24)
plt.figure(figsize=(12,8))

plt.subplot(3,1,1)
plt.plot(hours, day.prices_buy, label="Price")
plt.ylabel("Price ($/MWh)")
plt.legend()

plt.subplot(3,1,2)
plt.step(hours, day.demand_kw, label="Load", where='mid')
plt.step(hours, output.import_kw, label="Import", where='mid')
if day.allow_export:
    plt.step(hours, output.export_kw, label="Export", where='mid')
plt.ylabel("Power (kW)")
plt.legend()

plt.subplot(3,1,3)
plt.step(hours, output.charge_kw, label="Charge", where='mid')
plt.step(hours, output.discharge_kw, label="Discharge", where='mid')
plt.plot(hours, np.array(output.soc)*battery_details.capacity_kwh, label="SOC (kWh)")
plt.xlabel("Hour")
plt.ylabel("kW / kWh")
plt.legend()

plt.tight_layout()
plt.show()
```
